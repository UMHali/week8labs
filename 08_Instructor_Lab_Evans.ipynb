```python
# load data
gh_teams_df = pd.read_csv("github_teams.csv")
gh_teams_df.info()  # quick inspect
```

```python
# select columns Team_type, human_work, and work_per_human
(gh_teams_df
 .filter(["Team_type", "human_work", "work_per_human"])  # select desired columns
 .head()  # display first n rows
)
```

```python
# select columns thate in letter t using regex
(gh_teams_df
 .filter(regex="t$")  # select desired columns
 .head()  # display first n rows
)
```

```python
# sort data in descending order by Team_size_class, human_work, and work_per_human
(gh_teams_df
 .sort_values(by = ["Team_size_class", "human_work", "work_per_human"], ascending = False)  # select desired columns and sort
 .head()  # display first n rows
)
```

```python
# select human-bot teams that have bot_members_count value greater than and equal to 3
(gh_teams_df
 .query("Team_type == 'human-bot' and bot_members_count >= 3")  # keep rows based on condition
 .head()  # display first n rows
)
```

```python
# find the human teams that are large and have human_gini value greater than and equal to .75
(gh_teams_df
 .query("Team_type == 'human' and Team_size_class == 'Large' and human_gini >= 0.75")  # keep rows based on condition
 .head()  # display first n rows
)
```

```python
# how many teams are in small and large category
(gh_teams_df
 .query("Team_size_class == 'Large' or Team_size_class == 'Small'")  # keep rows based on condition
 .shape[0]  # fetch first dimension
)
```

```python
# how many teams in small or large category with a human_gini less than and equal to .25
(gh_teams_df
 .query("(Team_size_class == 'Large' or Team_size_class == 'Small') and human_gini <= 0.25")  # keep rows based on condition
 .shape[0]  # fetch first dimension
)
```

```python
# how many human teams are in the Medium category
(gh_teams_df
 .query("Team_type == 'human' and Team_size_class == 'Medium'")  # keep rows based on condition
 .shape[0]  # fetch first dimension
)
```

```python
# save columns team_size_class and work_per_human as a new DataFrame
gh_team_work_df = (
    gh_teams_df
    .filter(["Team_size_class", "work_per_human"])  # select desired columns
    .head()  # display first n rows
)
gh_team_work_df.info() # quick inspect
```

```python
# rename these two columns in the new Dataframe. Change human_gini to work_inequality and eval_survival_day_median to issue_resolution_time
# update filter on new dataframe
gh_team_work_df = (
    gh_teams_df
    .filter(["Team_size_class", "work_per_human", "human_gini", "eval_survival_day_median"])  # select desired columns
    .head()  # display first n rows
)

# rename columns
gh_rename_df = (
    gh_team_work_df
    .rename(columns={
        "human_gini" : "work_inequality",
        "eval_survival_day_median" : "issue_resolution_time"})
)
gh_rename_df.head()  # display first n rows
```
